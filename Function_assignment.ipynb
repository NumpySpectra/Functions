**6. Explain the concept of generators in Python and how they are defined.**

**A generator** in Python is a special type of function that yields values one at a time, instead of returning them all at once. It saves memory and is useful when dealing with large data.

**How we Define a Generator:**

Use the yield keyword instead of return.

def count_up_to(n):

    i = 1
    while i <= n:
        yield i
        i += 1
**How to Use:**

gen = count_up_to(3)

print(next(gen))  # 1

print(next(gen))  # 2

print(next(gen))  # 3


**7. What are the advantages of using generators over regular functions?**

**# Advantages of Using Generators over Regular Functions:**

**1. Memory Efficient**

Generators yield one value at a time, so they don’t store the entire result in memory like regular functions with lists.

Good for handling large datasets.

**2. Lazy Evaluation**

Values are generated only when needed, not all at once.

Improves performance for long loops or big calculations.

**3. Easy to Implement**

Generators are simpler to write using yield instead of writing full iterator classes.

**4. Improves Speed in Some Cases**

Because of lazy evaluation, they can start working immediately without waiting to generate the full result.

Example:

def generate_numbers():

    for i in range(1, 1000000):
        yield i
This uses very little memory compared to returning a huge list.

 **10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

 **1. map()**

Applies a function to each item in an iterable.

Returns a new iterable with the transformed items.

**Example:**


nums = [1, 2, 3]

squared = map(lambda x: x*x, nums)

print(list(squared))  # [1, 4, 9]

**2. filter()**

Applies a function that returns True or False to each item.

Returns a new iterable with only the items where the function is True.

**Example:**


nums = [1, 2, 3, 4]

even = filter(lambda x: x % 2 == 0, nums)

print(list(even))  # [2, 4]

**3. reduce() (from functools module)**

Applies a function cumulatively to the items of an iterable, reducing it to a single value.

**Example**:


from functools import reduce

nums = [1, 2, 3, 4]

sum_all = reduce(lambda x, y: x + y, nums)

print(sum_all)  # 10

 **1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list**

```python
def sum_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total


nums = [1, 2, 3, 4, 5, 6]
print(sum_even_numbers(nums))
```

**4. Write a Python function that checks if a given number is prime or not from 1 to 200.**

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True


num = 17
if 1 <= num <= 200:
    print(f"{num} is prime?" , is_prime(num))
else:
    print("Number should be between 1 and 200.")
```

 **5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.**

```python
class Fibonacci:
    def __init__(self, terms):
        self.terms = terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.a

# Example
fib = Fibonacci(7)
for num in fib:
    print(num)
```

 **6. Write a generator function in Python that yields the powers of 2 up to a given exponent.**

```python
def powers_of_two(max_exp):
    for exp in range(max_exp + 1):
        yield 2 ** exp

# Example
for value in powers_of_two(5):
    print(value)
```

 **9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.**

```python
def c_to_f(celsius):
    return (celsius * 9/5) + 32

celsius_temps = [0, 20, 37, 100]

fahrenheit_temps = map(c_to_f, celsius_temps)

print(list(fahrenheit_temps))
```

 **10. Create a Python program that uses `filter()` to remove all the vowels from a given string.**

```python
def is_not_vowel(char):
    return char.lower() not in 'aeiou'

input_str = "Hello World"
filtered_chars = filter(is_not_vowel, input_str)
result = ''.join(filtered_chars)

print(result)
```
